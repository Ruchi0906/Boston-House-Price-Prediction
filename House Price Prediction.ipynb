```python
def Replace_nan(df):
    Q=pd.DataFrame(df.isna().sum(),columns=["Count_null"])
    for i in Q[Q.Count_null>0].index:
        if(df[i].dtypes == "object"):
            x = df[i].mode()[0]
            df[i] = df[i].fillna(x)
        else:
            y = round(df[i].mean(),4)
            df[i] = df[i].fillna(y)
```

# ANOVA

```python
def ANOVA(df,cat,con):
    from pandas import DataFrame
    from statsmodels.formula.api import ols
    rel = con + ' ~ ' + cat
    model = ols(rel,df).fit()
    from statsmodels.stats.anova import anova_lm
    anova_results = anova_lm(model)
    Q = DataFrame(anova_results)
    a = Q["PR(>F)"][cat]
    return round(a,4)
```

```python
for i in X.columns:
    if(X[i].dtypes == "object"):
        a = ANOVA(train,i,"SalePrice")
        print(i,a)
        print("------------------------------")
        if(a<0.05):
            imp_cols.append(i)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=31)
```

# Regression Models

    1. Ridge
    2. Lasso

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
def model(mobj):
    model = mobj.fit(xtrain,ytrain)
    from sklearn.metrics import mean_absolute_error
    tr_pred = model.predict(xtrain)
    tr_err = round(mean_absolute_error(ytrain,tr_pred),4)
    ts_pred = model.predict(xtest)
    ts_err = round(mean_absolute_error(ytest,ts_pred),4)
    return tr_err,ts_err,mobj
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
model(lr)
```

# Ridge

```python
rr = Ridge()
model(rr)
```

# Lasso

```python
ls = Lasso()
model(ls)
```

# Tune the alpha value

```python
def cv_tune(mobj):
    model = mobj.fit(xtrain,ytrain)
    tg = []
    x = 0.9
    for i in range(0,1000,1):
        x = x + 0.1
        x =round(x,3)
        tg.append(x)
    
    tp={"alpha" : tg}
    from sklearn.model_selection import GridSearchCV
    cv = GridSearchCV(mobj,tp,scoring="neg_mean_squared_error",cv=5)
    cvmodel = cv.fit(xtrain,ytrain)
    return cvmodel.best_params_
```

```python
cv_tune(rr)  #alpha value for ridge
```

```python
cv_tune(ls)   #alpha value for lasso
```
